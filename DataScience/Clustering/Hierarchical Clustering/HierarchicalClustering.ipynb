# Hierarchical Clustering

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Datas

```python
datas = pd.read_csv('datas.csv')
```

### Take variable

```python
X = datas.iloc[:, 3:].values
```

### Use Hierarchical


```python
from sklearn import cluster
from sklearn.cluster import AgglomerativeClustering

clusterer = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')  
prediction = clusterer.fit_predict(X)

print(prediction)
```

### Visualize it

```python
plt.scatter(X[prediction==0,0],X[prediction==0,1],c = 'red')
plt.scatter(X[prediction==1,0],X[prediction==1,1],c = 'green')
plt.scatter(X[prediction==2,0],X[prediction==2,1],c = 'blue')
plt.scatter(X[prediction==3,0],X[prediction==3,1],c = 'black')
```

### Visualize Dendrogram

```python
import scipy.cluster.hierarchy as sch 
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
```
