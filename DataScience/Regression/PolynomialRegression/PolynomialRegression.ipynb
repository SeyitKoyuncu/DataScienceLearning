# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read Datas From CSV

```python
datas = pd.read_csv('Maaslar.csv') #read csv file with pandas librarys
```

## Seperate Datas

```python
x = datas.iloc[:, 1:2]
y = datas.iloc[:, -1]

X = x.values
Y = y.values

print(X)
print(Y)
```

## Show that Linear Regresssion Is Wrong Model For Polynomials Datas

```python
from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
linReg.fit(x,y)

plt.scatter(x,y, color = 'red')
plt.plot(x,linReg.predict(x), color = 'blue')

```

## Use Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

PolyReg = PolynomialFeatures(degree = 2) #say that create 2nd degree polynomial feature object

#transform datas to polynomial datas
x_poly = PolyReg.fit_transform(X) #formula of poltynomial regression = B*x^0 + B1*x^1 +B2*x^2+ B3*x^3... Frist column of x_poly x^0, second x^1, third x^2 

print(x_poly)
linReg2 = LinearRegression()
linReg2.fit(x_poly,y)
plt.scatter(x,y, color = 'red')
plt.plot(x,linReg2.predict(x_poly), color = 'blue')
plt.show()
```

## In this dataset larger degreees is better for predict 

```python
PolyReg = PolynomialFeatures(degree = 5)
x_poly = PolyReg.fit_transform(X)  
print(x_poly)
linReg2.fit(x_poly,y)
plt.scatter(x,y, color = 'red')
plt.plot(x,linReg2.predict(x_poly), color = 'blue')
```
