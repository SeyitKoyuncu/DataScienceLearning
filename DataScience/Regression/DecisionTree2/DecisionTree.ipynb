## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# Training the Decision Tree Regression Model on the Whole Dataset

```python
from sklearn.tree import DecisionTreeRegressor

dec_tree = DecisionTreeRegressor(random_state = 0)
dec_tree.fit(x,y)
```

# Predicting a new result

```python
y_predict = dec_tree.predict([[4.5]])
print(y_predict)
```

# Visulazing the Decision Tree Regression Results (higher resolution)

```python
#If you have more than one features you can use lower resolution, this statement for one featrue ' s datasets

X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(x,y)
plt.plot(X_grid,dec_tree.predict(X_grid), color = "red")
plt.title("Decesion Tree Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary Label")
plt.show()
```

# Visulazing the Decision Tree Regression Results (lower resolution)

```python
plt.scatter(x,y)
plt.plot(x,dec_tree.predict(x), color = "red")
plt.title("Decesion Tree Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary Label")
plt.show()
```
