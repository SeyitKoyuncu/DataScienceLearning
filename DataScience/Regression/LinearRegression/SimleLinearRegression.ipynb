## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Training the Simple Linear Regression model on the Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #create LinearRegression object
regressor.fit(X_train,y_train)
```

Predicting the Test set results

```python
y_pred = regressor.predict(X_test) #y_test contains the real salarys and y_pred is containing the predicted salaries
```

## Visualising the Training set Results

```python
plt.scatter(X_train, y_train, color = 'red') #drawing red points for real salary and experience datas in the training set
plt.plot(X_test, regressor.predict(X_test), color = 'blue') #drawing blue graph for real experience and predicted salarys these can be x_train or x_test
# doesnt matter.
plt.title('Salary vs Experience(Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red') #drawing red points for real salary and experience datas in the test set
plt.plot(X_train, regressor.predict(X_train), color = 'blue') #drawing blue graph for real experience and predicted salarys
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
