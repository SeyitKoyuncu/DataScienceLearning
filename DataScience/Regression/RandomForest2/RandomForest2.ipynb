## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# Training the Random Forest Regression Model on the Whole Dataset

```python
from sklearn.ensemble import RandomForestRegressor

#n_estimator = how much tree be used
forest_tree = RandomForestRegressor(n_estimators=5000, random_state = 0)
forest_tree.fit(x,y)
```

# Predicting a new result

```python
y_predict = forest_tree.predict([[4]])
y_predict2 = forest_tree.predict([[4.5]])

print(y_predict)
print(y_predict2)
```

# Visulazing the Random Forest Regression Results (higher resolution)

```python
#If you have more than one features you can use lower resolution, this statement for one featrue ' s datasets
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(x,y)
plt.plot(X_grid,forest_tree.predict(X_grid), color = "red")
plt.title("Decesion Tree Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary Label")
plt.show()
```

# Visulazing the Random Forest Regression Results (lower resolution)

```python
plt.scatter(x,y)
plt.plot(x,forest_tree.predict(x), color = "red")
plt.title("Decesion Tree Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary Label")
plt.show()
```
